```python
# task 6
# Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.
value=str(input("enter column name:"))
def median(value):
  for i in housing.columns:
    if i==value:
      return housing[value].median()
print(f"the median of {value} is:",{housing[value].median()})

median(value)

```

```python
# task 10
# Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
 # If the total bedrooms is 11 or more but less than 1000,it should be medium,
 #  otherwise it should be considered large.
def category(size):
  if size['total_bedrooms']<=10:
    return 'Small'
  elif size['total_bedrooms']>=11 and size['total_bedrooms']<1000:
    return 'medium'
  elif size ['total_bedrooms']>=1000:
    return 'large'
```

```python
housing['total_bedroom_size'] = housing.apply(lambda size: category(size),axis=1)
housing['total_bedroom_size']
housing
```
